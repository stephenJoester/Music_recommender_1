## Import & initiate

```python
from scraper import Request, Scraper
import os 
import pandas as pd
from fp.fp import FreeProxy
```

```python
DATASET_PATH = 'data/'
AUDIO_PATH = os.path.join(DATASET_PATH + 'audio') 
```

```python
from datetime import datetime

current_date = datetime.now().strftime("%d_%m")
print("Current date:", current_date)

df_spotify = pd.read_csv(DATASET_PATH + f"tracks_metadata_{current_date}.csv") 
```

```python
from dotenv import load_dotenv
import os

load_dotenv() 

lastfm_api_key = os.environ['LASTFM_API_KEY']

request = Request().request()
scraper = Scraper(session=request, lastfm_api_key=lastfm_api_key)
```

## Using Last.fm API

```python
df_tags_api = scraper.get_lastfm_tags_df(df_spotify)  
```

```python
df_spotify.loc[df_spotify['track_id'].isin(df_spotify['track_id']), :] = df_tags_api
df_spotify
```

```python
df_spotify.to_csv(DATASET_PATH+f"tracks_metadata_with_tags_{current_date}.csv",index=False)
```

## Fix untagged rows

```python
merged_df = pd.read_csv(DATASET_PATH+f"tracks_metadata_with_tags_{current_date}.csv")
untagged_df = merged_df.loc[merged_df['tags'].apply(lambda x: eval(x) if isinstance(x, str) else x).apply(lambda x: len(x) < 1)]
untagged_df
```

```python
previous_proxy = []
fixxed_df_whole = pd.DataFrame()
previous_untagged_df = pd.DataFrame()
while len(untagged_df) != 0: 
    request = Request().request()
    scraper = Scraper(session=request, lastfm_api_key=lastfm_api_key)
    
    # proxy rotation 
    proxy = {"https": FreeProxy(timeout=1,https=True).get()}
    if (proxy in previous_proxy) : 
        continue
    previous_proxy.append(proxy)
    print(proxy)
    
    # crawl
    try : 
        fixxed_df = scraper.fix_untagged_df(untagged_df,proxy=proxy)
        # merge
        merged_df.loc[merged_df['track_id'].isin(fixxed_df['track_id']), :] = fixxed_df
        try : 
            untagged_df = merged_df.loc[merged_df['tags'].apply(lambda x: eval(x) if isinstance(x, str) else x).apply(lambda x: len(x) < 1)]
            # untagged_df = untagged_df[~(untagged_df['track_id'].isin(fixxed_df_whole[fixxed_df['tags'] == '[]']['track_id']))]
            print(len(untagged_df))
        except :
            raise
        
    except :
        continue
    
    # previous_untagged_df = untagged_df.copy()
    
    
    
```
